```python
import time
```

```python
print("-----------WELCOME TO ABC BANK-----------\n\n")
print("Insert Your Card")

print("------------------------------------------")

time.sleep(3)

password = 0000

Pin =int(input("Enter your Four Digit PIN\n"))
print("------------------------------------------\n")

Balance = 50000

if Pin == password:
    while True:

        print("****BANKING OPTIONS****\n")
        print("Press" "1 == To Check Current Balance\n")
        print("Press" "2 == To Deposit Cash\n")
        print("Press" "3 == To Withdraw Cash\n")
        print("Press" "4 == Exit\n")
        print("------------------------------------------\n")
        try:
            Option = int(input("Please Press Button\n"))
            print("------------------------------------------\n")
        except:
            print("Please enter Valid option")

        if Option == 1:
            print(f"Your Current Balance is : {Balance} Rupees")
            print("------------------------------------------\n")
        if Option == 2:
            Deposit = int(input("Please Enter Deposit Amount\n"))
            Balance += Deposit
            print("------------------------------------------\n")
            print(f"{Deposit} Rupees is Deposited to Your Account\n")
            print(f"Current Balance is : {Balance} Rupees\n")
            print("------------------------------------------\n")
        if Option == 3:
            Withdrawal = int(input("Please Enter Withdraw Amount\n"))
            Balance -= Withdrawal
            print("------------------------------------------\n")
            print(f"{Withdrawal} Rupees is Debited from Your Account\n")
            print(f"Current Balance is : {Balance} Rupees\n")
            print("------------------------------------------\n")
        if Option == 4:
            print("Thanks for Reaching Us\nHave a Nice Day!")
            break

else:
    print("Incorrect Pin!\nPlease Try Again")
```
